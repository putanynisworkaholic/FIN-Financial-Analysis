```python
import requests
import pandas as pd

# Define your Alpha Vantage API key
api_key = 'A5K2F1KTMVFXWP13'

# Define the symbol for the stock
symbol = 'FDX'

# Define the base URL for Alpha Vantage
base_url = 'https://www.alphavantage.co/query'

# Function to fetch financial data
def fetch_data(function, symbol, api_key):
    params = {
        'function': function,
        'symbol': symbol,
        'apikey': api_key
    }
    response = requests.get(base_url, params=params)
    data = response.json()
    return data

```

```python
# Fetch income statements (yearly and quarterly)
income_yearly = fetch_data('INCOME_STATEMENT', symbol, api_key)['annualReports']
income_quarterly = fetch_data('INCOME_STATEMENT', symbol, api_key)['quarterlyReports']

# Fetch balance sheets (yearly and quarterly)
balance_yearly = fetch_data('BALANCE_SHEET', symbol, api_key)['annualReports']
balance_quarterly = fetch_data('BALANCE_SHEET', symbol, api_key)['quarterlyReports']

# Fetch cash flow statements (yearly and quarterly)
cash_flow_yearly = fetch_data('CASH_FLOW', symbol, api_key)['annualReports']
cash_flow_quarterly = fetch_data('CASH_FLOW', symbol, api_key)['quarterlyReports']

# Convert the data to pandas DataFrames
df_income_yearly = pd.DataFrame(income_yearly)
df_income_quarterly = pd.DataFrame(income_quarterly)

df_balance_yearly = pd.DataFrame(balance_yearly)
df_balance_quarterly = pd.DataFrame(balance_quarterly)

df_cash_flow_yearly = pd.DataFrame(cash_flow_yearly)
df_cash_flow_quarterly = pd.DataFrame(cash_flow_quarterly)

# Display the dataframes
print(df_income_yearly.head())
print(df_income_quarterly.head())
print(df_balance_yearly.head())
print(df_balance_quarterly.head())
print(df_cash_flow_yearly.head())
print(df_cash_flow_quarterly.head())
```

```python
def clean_and_convert_to_numeric(df):
    # Convert all columns except the index to numeric, replacing non-numeric entries (e.g., None) with 0
    df = df.apply(pd.to_numeric, errors='coerce').fillna(0)
    return df
```
